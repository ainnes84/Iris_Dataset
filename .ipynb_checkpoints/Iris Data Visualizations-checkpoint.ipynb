```python
# For this project I have compied Ben Hamner's Python Data Visualizations from the
# Iris Dataset on Kaggle

# To start we must first import pandas- this is a data processing and CSV file I/O library
import pandas as pd
```

```python
# Next we will import seaborn which is a Python graphing library.
import warnings # This part of the code ignored the warnings that seaborn generates
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt # Finally we import MatPlotLib in order to show the graphs
sns.set(style="white", color_codes=True)
```

```python
# Our next step is to load the Iris Flower Dataset
# This piece of code also puts the Iris Dataset into a Pandas DataFrame
iris=pd.read_csv("Iris.csv")
```

```python
# Finally we want to see what's in the Iris Data.  This will be the final step before
# starting to make graphs.  In order to execute this cell you must press shift+enter
# This displays the table that you will see below that shows Sepal Length, Sepal Width,
# Petal Length, and Petal Width.  All these factors can actually determine which species
# and flower that is being described.
iris.head()
```

```python
# Let's see how many examples we have of each species
# As you can see below each species has a total of 50 examples each.
iris["Species"].value_counts()
```

```python
# Ourfirst plot is using the .plot extension from Pandas Dataframes
# We'll use this to make a scatterplot of the Iris features.
# Which for this example we will use Sepal Length and Sepal Width in cm,
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
plt.show() #This is not needed in kaggle but in order to show the graph in Python
# you must use the plt.show() command.  This will be used many more times throughout
# this project
```

```python
# We can also use the seaborn library to make a similar plot
# A seaborn jointplot shows bivariate scatterplots and unicariate histograms
# in the same figure
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=7) 
# The joint plot makes it easier to see what the most common Sepal Width and
# Sepal Height are.  Changing the size will either make the graph larger or smaller.
# The orginal size was 5 but I changed it to 7 to spread the points out a bit more
# anddd make it easier to read.
plt.show()
```

```python
# One piece of information is missing in the plots above is what species each
#plant is
# We'll use seaborn's FacetGrid to color the scatterplot by species
sns.FacetGrid(iris, hue="Species", size=7) \
.map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
.add_legend()
plt.show() #Just like the previous chart I increased the size to make the points 
# more clear.
```

```python
# We can look at an individual feature in Seaborn through a boxplot
# In this case we are looking at Petal Length.
# As you can see the Iris-Virginica specieas has the longer petal lengths but
# also the furthest outliers.
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
plt.show()
```

```python
# One way we can extend this plot is adding a layer of individual points to it
# through Seaborn's striplot

# We'll use jitter=True so that all the points don't fall in a single vertical lines
# above the species.  This makes the plot a little more clear as you can see below.

# Saving the resulting axes as ax each time causes the resulting plot to be shown
# on top of previous axes
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True,
                  edgecolor="gray")
plt.show()
```

```python
# A violin plot combines the benefits of the previous two plots and simplifies them
# These types of plots basically take a boxplot and make it easier to see where
# the data lies instead of keeping it boxed in the middle and identifying only the farthest
# outliers.
# Denser regions of the data are fatter, and sparser thiner in a violin plot
sns.violinplot(x="Species", y="PetalLengthCm", data=iris, size=7)
plt.show()
```

```python
# A final seaborn plot useful for looking at unicariate relations is the kdeplot,
# which creates and visualizes a kernel density estimate of the underlying feature
sns.FacetGrid(iris, hue="Species", size=6) \
.map(sns.kdeplot, "PetalLengthCm") \
.add_legend()
plt.show()
```

```python
#Another useful seaborn plot is the pairplot, which shows the bivariate relation
#between each of features

# From the pairplot, we'll see that the Iris-setosa species is seperated from the other
# two across all feature combinations
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)
plt.show() # I tried increasing the size to this plot but it became harder to read.
# Also decreasing the size made the data look cluttered.
```

```python
# The diagonal elements in a pairplot show the histogram by default but
# we can update these elements to show other things such as a kde.
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3, diag_kind="kde")
plt.show() #As you can see when you use a kde it gets rid of the kde and shows the
# distribution more clearly.
```

```python
# Now we are going back to Pandas. We can make a boxplot with Pandas
# with each feature showed specifically by each species.
# These boxplots make it easier to see which plants have both longer and wider, 
# Petals and Sepals.  As you can see the Iris-Virginica has almost all other species
# beat besides in Sepal Width.
iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
plt.show()
```

```python
# This is a technique in pandas that is called Andrews Curves.
# It i a way to visualize structure in high-demensional data.
from pandas.tools.plotting import andrews_curves
andrews_curves(iris.drop("Id", axis=1), "Species")
plt.show()
```

```python
# Another multivariate visualization technique in pandas is parallel coordinates.
# This plots each feature on a seperate column then draws lines connecting
# the features for each data sample.  The previous Andrews Curves plot is more
# of a smoothed out version of this plot.
from pandas.tools.plotting import parallel_coordinates
parallel_coordinates(iris.drop("Id", axis=1), "Species")
plt.show()
```

```python
# Our final plot and multivariate visualization tool in pandas os radviz.
# This puts each feature as a point on a 2D plane. Which then simulates having each
# sample attached to those points by the relative value for that feature.
# As you can see the Iris-Setosa is completed seperated from the other two species.
from pandas.tools.plotting import radviz
radviz(iris.drop("Id", axis=1), "Species")
plt.show()
```
